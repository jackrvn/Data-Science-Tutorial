## Preparing our dataset for analysis
Now that we have a clearer idea of the type of information thats available, in this dataset, before we can actually doo anything with it, we need to make sure that it is ready to actuall be worked with.

### Dropping columns that aren't needed
The first thing we want to do is try and figure out what information we *don't* need, as we will not be looking at them.

Normally you would have to figure this part out, but as this is a tutorial I can do that for you. In this instance, we will not be using the column "key" in our analysis as we can get a pretty good understanding of our data without it. We also won't bee using 'release_date' as we are given the year, which will be enough for our analysis. So we are going to go ahead and 'drop' both of these from our dataset. To do this, we can use **.drop()** on our dataset and pass in two arguments:
* **A list of column names** we want to drop
* The 'axis' value which describes whether we want to drop from the columns or the rows (because we want to drop columns, we have to use **axis=1** since as standard axis is set to 0 meaning rows).

```python
spotify_dataset = spotify_dataset.drop(['key', 'release_date'], axis=1)
spotify_dataset.info()
```

Now we can see that the number of columns has gone down to 17 from 19, and the columns 'key' and 'release_date' are no longer in our dataset - great!

In case you are concerned - dropping a column does not remove the data permananently from the original file. If you were to go back and run the original cell which read in our data again, you would get the 'key' column back! Because we set the results of our drop to the variable spotify_dataset, it replaces the original dataset that it contained.

If you wanted to however, you could always change it so that the version with the dropped columns was stored in a different variable (but this is not necessary in most cases)

### Converting 'duration' into a more readable format
Another way we can prepare our data, is by converting the type of values  certain columns may hold into ones that are a little more easy for us to understand.

If you look at the previous cell containing the datasets info, we can see that the column 'duration_ms' is currently in the form of milliseconds... which isn't super helpful for us to understand (unless you're someone who tells the time in ms, in which case go you!)

So to resolve this, we're going to go ahead and use a function called **pd.to_timedelta()** which will create a new column with actual time values for us! To do this, we will pass two arguments:
* The **column we want to convert** (you can easily specify a column by using dataset_name['column_name'], so in our case this will be spotify_dataset['duration_ms']
* The **unit** of measurement, in this case we can use **unit='ms'**

Now, we don't want to change the original column "duration_ms" in any way, as this will be helpful to us later down the line when we want to do do some maths on this data. So we will save the results into a new column labelled "duration" in our dataset. We can do this super easily be simply setting the results to **spotify_dataset['duration']**

```python
spotify_dataset['duration'] = pd.to_timedelta(spotify_dataset['duration_ms'], unit='ms')
spotify_dataset.info()
```

As you can see, we now have a new 18th column called 'duration' of the type 'timedelta', and the column "duration_ms" has not been changed.

Now, let's have a look at exactly what this column looks like. To do this, we can use **spotify_dataset['duration'].head()** to get the first 5 results from the column 'duration' in our dataset:

### Okay... So what can we do then when we have a lot of data like this?
Rather than trying to look at every single song duration in 2020, we can simply take the **average** of this data. If you're unfamiliar with the concept of an average, you can think of it being like **the most typical value in a set of data.** Being able to find out what this number is really helpful for us, when we are working with larger sets of data.

There's a few ways that we can describe the average, but the main one you'll often see someone using the **mean**. You probably already know what this is, but in case you don't remember, we find the mean of a dataset by **adding all of the values in our data together, and then dividing by the total number of values we have in our data.** So for example, if we had a list my_list = [1, 2, 3]; to find the mean for my_list we would just do  $\frac{1+2+3}{3}$, meaning our mean for my_list = 2.

Luckily, pandas has a pretty simple way for finding out the mean of a whole column, the **.mean()** function!

Looking back at the **songs2020** variable we have made, we can go ahead and grab the mean for the duration of all the songs published in 2020. Let's save this as a variable called **songs2020_duration_ms_mean**.

#### But wait!!
Remember how we chose to save both **duration** as a timedelta datatype, as well as **duration_ms** which stores how many milliseconds long a song is? This is where having both comes in handy! Since we want to **plot our data**, meaning we convert it into graphs, we want to make sure that the data we are working with **is an integer**, so that we can do this easily. For this reason, **we want to store the mean of duration_ms, not duration:** 

```python
songs2020_duration_ms_mean = songs2020['duration_ms'].mean()
songs2020_duration_ms_mean
```

The above output may not seem super nice to read for us, so if you want to know what the mean duration of all of the songs from 2020 in this dataset are, we can just run the .mean() function on the normal duration column and print that out too (we don't need to store it anywhere however, as we will not be using it again).

```python
songs2020['duration'].mean()
```

### Task 1: "Smile like you mean it"
Now that we have the average song length for the year 2020, let's grab the ones for the years 2019 and 2018 too so that we can analyse the song length for those three years.

For your first real task, you will need to fill in the variables **songs2019_duration_ms_mean** and **songs2018_duration_ms_mean** to contain the mean for the duration **(in ms)** for all the songs from the years 2019 and 2018, respectively. 

```python
''' Remember, don't alter our original variable spotify_dataset! 
Instead you can use these two variables below to hold altered versions
of the main dataset.
'''
songs2019 = spotify_dataset[spotify_dataset['year']==2019]
songs2018 = spotify_dataset[spotify_dataset['year']==2018]

songs2019_duration_ms_mean = songs2019['duration_ms'].mean()
songs2018_duration_ms_mean = songs2018['duration_ms'].mean()
```

Well done! Now that we have our 3 variables, lets make our first ever graph!

Since we're only really dealing with 3 values, we can plot our means into a bar chart using a function called **plt.bar()** (remember, 'plt' is just the nickname we gave to matplotlib). To do this, we will want our **x-axis to denote each year** (so 2020, 2019 and 2018); and our **y-axis as the means we have just calculated** (note: bar charts in matplotlib use the parameter **height** to refer to the y-axis).

We can put these values into two lists, **x_years** and **x_means** (making sure both lists are in the same chronological order), so that matplotlib knows exactly what values it needs to use to make the barchart. Note, we will also need to make sure that our years are stored in our x_years list as **strings, not integers** so that matplotlib knows that we want to use these as **static** labels rather than numbers.

Finally, we want to also label our axes and give our graph a title. While these are necessary, it's good practice to get into the habit of doing, so that other people (as well as you in the future) can understand exactly what the graph is representing. In order to make the graph a little more readable too, we can also provide a list of different colours to tell matplotlib what colour each bar should be.

```python
x_years = ['2018', '2019', '2020']
y_means = [songs2018_duration_ms_mean, songs2019_duration_ms_mean, songs2020_duration_ms_mean]

plt.bar(x=x_years, height=y_means, color=['pink', 'purple', 'palevioletred'])
plt.xlabel('Year')
plt.ylabel('Mean Track Duration (ms)')
plt.title('Mean Track Duration Over the Last 3 Years')
plt.show()
```

Wow! Just like that, you've made your first ever graph, AND it's super pretty (if I do say so myself). Well done! Now, let's do some analysis!

This way, it makes it a lot easier for us as data scientists to see if we can spot any **trends** in our data- and just to be clear, I'm not talking about TikTok trends. When we talk about a trend in datascience, we mean any patterns or a particular direction our data might be moving.  If you're still not sure what I mean, don't worry, it will become pretty clear in a bit!

Now since we would have to calculate 10 years worth of data to be able to create this graph, this can get pretty tedious and tiring to write out the same line of code 10 times. So, lets automate it!

If you have some experience in python, or any programming language really, then you are likely already familiar with the concept of a **for loop** - so we're gonna use one of those to calculate the mean song duration for each year, for us!

We'll start by making a list which contains all the years from the last 10 years (so in my case, this is 2010-2020). While we could write out every single year out by hand and put it into a list, we can instead use the **range(start,end)** function to do this for us! We will then store this as the variable **last10years**. We should note, since a line plot is a **continuous** plot, instead of a static plot (meaning the graph doesn't have any breaks in it), this time we want our years to remain as **integers**, not strings. This also means that we can use the list **last10years** for our x axis too! Another important thing to remember also, is that the range function is **exclusive, not inclusive**, meaning we will need to actually do **range(2010, 2021)** in order to get a list from 2010 to 2020 (otherwise it will only go to 2019!).

Like we did with the barchart, we will also want to make another list **last10years_duration_means** which will contain all the song duration means from the last 10 years.

Our for loop will then loop through each year in **last10years**, and grab the mean for that year in a similar way to what we did before. However, this time we want to also make sure that we are appending the resulting mean in our list **last10years_duration_means**.

Finally, we can use **plt.plot(x,y)** to create our line plot, where x = **last10years** and y = **last10years_duration_means**.

```python
last10years = range(2010,2021)

last10years_duration_means = []

for i in last10years:
    
    current_year = spotify_dataset[spotify_dataset['year']==i]
    current_year_mean = current_year['duration_ms'].mean()
    
    last10years_duration_means.append(current_year_mean)
    
plt.plot(last10years, last10years_duration_means)
plt.xlabel('Year')
plt.ylabel('Mean Track Duration (ms)')
plt.title('Mean Track Duration Over the Last 10 Years')
plt.show()
```

Yay! We've got our graph! Before we do our analysis, lets also quickly go ahead and print out a readable version of the **mean song length** for both the years **2010 and 2020**, so we can have a little more detail.

So far we've used Matplotlib to create bar charts and scatter graphs to provide some basic insights into our data, but the library is capable of so much more!

For example, we can get a better idea of the distribution of our data by creating something called a histogram. This is a statistical tool that categorises our data into a number of containers known as "bins", with each bin representing a certain range of values. A spike in this kind of graph means there are a lot of values close to that point.

```python
plt.hist(spotify_dataset['duration_ms'] // 60000, bins=90)
plt.xlabel('Duration (minutes)')
plt.ylabel('No. of songs')
plt.title('Song Length Distribution')
plt.show()
```

In this example, we can see how the duration of the songs in our dataset is heavily concentrated around the 2-5 minute mark, with very few songs reaching over 20 minutes long.

Histograms are incredibly useful in data science and statistics, we can use them to compare distributions and help to verify or challenge the assumptions we are making about our data (for example if there are any outliers that might be skewing our results). Unlike a bar or scatter plot, we only need one variable to create a histogram. However, in Matplotlib, we can overlap histograms to allow for easy comparisons between multiple distributions.

The number of bins we use decides the number of columns our values are divided into, a higher number of bins effectively adds more detail to our visualisation.
